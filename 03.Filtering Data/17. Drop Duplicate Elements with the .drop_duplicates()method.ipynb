```python
import pandas as pd
df_laptops= pd.read_csv('laptop_data.csv')
df_laptops.head()
```

# 1 drop_duplicates(0)

```python
# drop duplicates inn 2 or moree columns
df_laptops.drop_duplicates(['Company'])
```

```python
# sort dataframe ascending by company and price
# cheapest first and most expensiove last
df_laptops.sort_values(['Company','Price'])
```

```python
# cheapest : keep ='first'
df_laptops.drop_duplicates(['Company'], keep='first')[['Company','Price']]
```

```python
# expensive : keep ='last'
df_laptops.drop_duplicates(['Company'], keep='last', inplace=True, ignore_index=True)

df_laptops[['Company','Price']]
```
