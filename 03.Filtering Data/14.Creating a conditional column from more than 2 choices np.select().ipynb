# 1 Creating a conditional column from more than 2 choices :np.select()

```python
import numpy as np
```

```python
# create an array based on multiple tiers price tiers(+2 choices)
# create "condition" and "values" variables 
Conditions = [
        df_laptops['Price']>80000,
        (df_laptops['Price']>70000) & (df_laptops['Price']<=80000),
        (df_laptops['Price']>60000) & (df_laptops['Price']<=70000),
        df_laptops['Price']<=50000

]

Values=['Too Expensive','Expensive','Affordable','Cheap']
```

```python
# set it to an new column
df_laptops['Price_tiers'] =np.select(Conditions, Values, default='Unknown')
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "price_tier" column
df_laptops['Price_tiers'].value_counts()
```

# 2 EXERCISE 

```python
# create an array based on multiple screen size tiers (+2 choices)
# too big>16 , big>14 , small >12, too small<12

# create "conditions" and "values" variable
contions=[
    df_laptops['Inches']>16,
        (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
        (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
        df_laptops['Inches']<=12
]
values=['Too Big','Big','Small','Too small']
```

```python
# set it to new column
df_laptops["Screen_size"]=np.select(contions,values,default='Unknown')
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "screen_size"
df_laptops['Screen_size'].value_counts()
```
