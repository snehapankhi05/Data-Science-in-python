# 1 Create a column based on 1 conditions np.where()

```python
import numpy as np
```

```python
# create an aray based on price tiers
np.where(df_laptops['Price']>40000, 'Expensive','Cheap')
```

```python
# set it to a new column
df_laptops['Price_tier']=np.where(df_laptops['Price']>40000, 'Expensive','Cheap')
```

```python
# count values in price_tier column
df_laptops.value_counts('Price_tier')
```
