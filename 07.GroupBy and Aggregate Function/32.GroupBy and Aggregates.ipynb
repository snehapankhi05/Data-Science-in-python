# 1 agg()

```python
# calculate the sum of columns
df.agg('sum')
```

```python
# calculate the mean of numeric columns
# Only apply 'mean' to numeric columns
df.select_dtypes(include='number').agg('mean')
```

```python
# calculate the mean and count values
# Only for numeric columns
df.select_dtypes(include='number').agg(['mean','count'])
```

```python
# different aggreations per column: calculate the sum and mean of "sales_in_thousands" and sum and max of 'price_in_thousands'
df.agg({'Sales_in_thousands':['sum','mean'],
       'Price_in_thousands':['sum','max']})
```

```python
# aggregate over the columns
df[['Sales_in_thousands','Price_in_thousands']].agg('sum',axis=1)
```

```python
# equivalent (NaN behaves differently)
df['Sales_in_thousands']+df['Price_in_thousands']
```

```python
# aggregate different functions and rename the index of the resulting dataframe
df.agg(x=('Sales_in_thousands','sum'),
       y =('Price_in_thousands','sum'))
```

# 2 The Split-Apply-Combine Strategy

```python
# split data into separate groups
pass_filter=df['Vehicle_type']=="Passenger"
car_filter=df['Vehicle_type']=="Car"
df
#df['Vehicle_type'].value_counts()
```

```python
# apply an operation
pass_avg=df[pass_filter]['Sales_in_thousands'].mean()
car_avg=df[car_filter]['Sales_in_thousands'].mean()
```

```python
# combine results
pd.DataFrame({'Vehicle_type':['Car','Passenger'],
             'Sales_in_thousands':[car_avg,pass_avg]})
```

# 4 groupby() and agg()

```python
# find the minimum and maximum values on each column
df.groupby('Vehicle_type').agg(['min','max'])
```

```python
# set a name for aggregated column
# find the min Engine_size and max Horsepower
df.groupby('Vehicle_type').agg(min_engine_size=('Engine_size','min')
                               ,max_Horsepower=('Horsepower','max'))
```

```python
# calculate the sum of Sales_in_thousands and mean of Price_in_thousands
df.groupby('Manufacturer').agg(sum_sales=('Sales_in_thousands','sum'),
                              mean_price=('Price_in_thousands','mean'))
```

```python
df.groupby('Manufacturer').sum()['Sales_in_thousands']
```

# Lambda
Apply your own function

```python
df.groupby('Manufacturer').sum()
```

```python
# get actual sales and price values
df.groupby('Manufacturer').sum().apply(lambda x:x*1000)[['Price_in_thousands','Sales_in_thousands']]
```

```python
# subtrack values fron the mean of each group
numeric_cols = df.select_dtypes('number').columns

# Subtract the mean of each group for numeric columns, keep all columns intact
df.loc[:, numeric_cols] = df.groupby('Manufacturer', group_keys=False)[numeric_cols].transform(
    lambda x: x - x.mean()
)
df
```
