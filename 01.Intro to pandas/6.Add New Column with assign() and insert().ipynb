# 1 Adding a new column
## 1.1 assign()

#### When to use assign?
#####   - Add multiple columns in single line of code
#####   -When you need to overwrite the values of an existing columns (best pratice)

#### It returns a new object (a copy) with all the originals columns in addition to the new one

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1,100,size=50)
score2 = np.random.randint(1,100,size=50)
```

```python
# create a series using the random numbers
series1=pd.Series(score1, index=np.arange(0,50))
series2=pd.Series(score2, index=np.arange(0,50))
```

```python
# using assign() to add multiple columns
df_exams=df_exams.assign(score1=series1,score2=series2)
```

```python
#using dataframe
df_exams
```

## 1.2 insert()

#### Inserts a new column at a specific position or index.

```python
# using index() to add a column at a specific position or index
df_exams.insert(1,'rank1',series1)
```
