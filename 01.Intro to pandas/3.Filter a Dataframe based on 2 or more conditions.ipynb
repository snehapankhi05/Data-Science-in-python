```python
import pandas as pd
```

```python
df_laptops= pd.read_csv('laptop_data.csv')
df_laptops.head(3)
```

# 1 Filter dataframe based on multiple conditions

```python
# find apple laptops that cost more than 30000 and company is apple
(df_laptops['Company']=="Apple") &(df_laptops['Price']>30000)
```

## 2.1 & (and)

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company']=="Apple") &(df_laptops['Price']>30000)]
```

## 2.2 |(or)

```python
# find laptops apple or dell laptops
df_laptops['Company']=="Apple"
df_laptops['Company']=="Dell"
(df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")
```

```python
# filter dataframe based on multiple conditions
df_laptops[(df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")]
```

```python
df_laptops[(df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")].value_counts('Company')
```

```python
# find laptops from apple or dell that cost over 30000
df_laptops['Company']=="Apple"
df_laptops['Company']=="Dell"
df_laptops['Price']>30000

((df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")) & (df_laptops['Price']>30000)
```

```python
# filter dataframe based on multiple columns
df_laptops[((df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")) & (df_laptops['Price']>30000)]
```

```python
df_laptops[((df_laptops['Company']=="Apple") | (df_laptops['Company']=="Dell")) & (df_laptops['Price']>30000)].value_counts("Company")
```
