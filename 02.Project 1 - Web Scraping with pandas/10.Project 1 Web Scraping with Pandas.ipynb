# 2 Read .csv from Multiple URLs with Pandas

##### - https://www.football-data.co.uk/mmz4281/2122/E0.csv
##### - https://www.football-data.co.uk/mmz4281/2122/E1.csv

#### Link: root + season + league 

```python
# link structure 
"https://www.football-data.co.uk/mmz4281/" + "2122" +"/" +"E0"+".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple Leagues

##### - https://www.football-data.co.uk/mmz4281/2122/E0.csv
##### - https://www.football-data.co.uk/mmz4281/2122/E1.csv
##### - https://www.football-data.co.uk/mmz4281/2122/E2.csv
##### - https://www.football-data.co.uk/mmz4281/2122/E3.csv
##### - https://www.football-data.co.uk/mmz4281/2122/EC.csv


```python
# creating list of leagues
leagues=['E0','E1','E2','E3','EC']
frames=[]

# looping through league , read multiple csv and append it into a list
for league in leagues:
    df=pd.read_csv(root+'2122'+ '/'+ league+'.csv',encoding='unicode_escape')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st , 2nd and 3rd element
frames[0]
```

## 2.2 Multiple Seasons

```python
# creating list of leagues
leagues=['E0','E2','E3']
frames=[]

# looping through league , read multiple csv and append it into a list
for league in leagues:
    for season in range(15,21):
        df=pd.read_csv(root + str(season) + str(season+1) + '/'+ league+'.csv',encoding='unicode_escape')
        df.insert(1,'season',season)
        frames.append(df)
# loop through range of seasons and insert new column
```

```python
# total frames (6 seasons X 3 leagues)
len(frames)
```

```python
# show 1st , and last element(new season column was created)
frames[17]
```

# 3 Organize All The Data in a Dictionary

```python
# create dictionary with original name of league as key
dic_countries={
               'Spanish La Liga':'SP1','Spanish Segunda Division':'SP2',
               'German Bundesliga':'D1',
               'English Premier League':'E0','English Premier League 1':'E2','English Premier League 2':'E3'
              }
```

```python
dict_historical_data={}
# creating list of leagues
#leagues=['E0','E2','E3']


# looping through league , read multiple csv and append it into a list
for league in dic_countries:
    frames=[]
    for season in range(15,21):
        df=pd.read_csv(root + str(season) + str(season+1) + '/'+ dic_countries+'.csv')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat= pd.concat(frames)
    dict_historical_data[league]=df_concat
```
