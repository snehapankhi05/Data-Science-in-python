```python
import pandas as pd
df=pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df
```

```python
df=df[['long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality'	,'club']]
df
```

# 1 Selecting with a single value
iloc[row_position , column_position ]

```python
# get the height of L.Messi
df.iloc[0,3]
```

```python
# get the weight of Cristiano Ronaldo
df.iloc[1,4]
```

```python
# get all the rows inside the 'height_cm column'
df.iloc[:,3]
```

```python
# get all the columns that correspond to the index 'L.Messi'
df.iloc[0,:]
```

# 2 Selecting with a list of values

```python
# get all the data about L.Messi and Cristiano Ronaldo
df.iloc[[0,1]]
```

```python
# get the height of L.Messi and Cristiano Ronaldo
df.iloc[[0,1],3]
```

```python
# get the height of L.Messi and Cristiano Ronaldo
df.iloc[0,[3,4]]
```

```python
# get height anf weight of L.Messi and Cristiano Ronaldo
df.iloc[[0,1],[3,4]]
```

# 3 Selecting a range of data with a slice
start:stop(Note that the start is included but the stop is excluded)

```python
# slice column labels age/club
# player=['L.Messi','Cristiano Ronaldo']
players=[0,1]
df.iloc[players,1:7]
```

```python
# slice index labels (top1 to top10 player)
# columns =['age'	,'dob'	,'height_cm','weight_kg']
col=[1,2,3,4]

# get top1 and top10 player name
df.iloc[0:10,col]
```

# 4 Selecting with conditions
iloc cannot accept a boolean series, but only a boolean list. We have to use the lis() function to convert a series into a boolean list.

```python
# one condition : select player with height above 180 cm
# columns =['age'	,'dob'	,'height_cm','weight_kg']
cols=[1,2,3,4]

df.iloc[list(df['height_cm']>180), cols]
```

```python
# multiple condition : select player with height above 180 cm and from Argentina


df.iloc[list((df['height_cm']>180) & (df['nationality']>'Argentina')) ]
```
