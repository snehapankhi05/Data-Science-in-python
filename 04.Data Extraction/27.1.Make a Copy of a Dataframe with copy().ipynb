```python
import pandas as pd
df=pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df=df[['long_name',	'age',	'dob',	'height_cm',	'weight_kg',	'nationality'	,'club']]
df
```

# 1 copy()

### 1.1 deep=True

```python
# deep=True by default 
df_copy=df.copy()
```

```python
# update the value in original dataframe
df.loc['L. Messi','height_cm']=180
```

```python
# copy vs origianl dataframe
df
df_copy
```

### 1.2 deep=False

```python
# deep=False (Any changes to the data of the original will be reflected)
df_copy=df.copy(deep=False)
```

```python
# update value in original dataframe
df.loc['L. Messi','height_cm']=180
```

```python
# shallow copy vs original dataframe


df_copy
```

### 1.3 Simple assignment


```python
# make a copy
df_copy_new=df
```

```python
# update value in original dataframe
df.loc['J. Oblak','height_cm']=180
```

```python
# copy vs original dataframe
df_copy_new
```
