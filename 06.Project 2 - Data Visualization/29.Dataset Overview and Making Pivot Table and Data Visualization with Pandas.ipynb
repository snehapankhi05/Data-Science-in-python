```python
import pandas as pd
```

```python
df=pd.read_csv('population_total.csv')
df
```

# 1 Making a Pivot table

.pivot() : returns reshaped Dataframe organized by given index/column values("pivot without aggregation")

```python
# dropping null values
df=df.dropna()
df
```

```python
# making a pivot table
df_pivot=df.pivot(index="year", columns="country",values="population")
```

```python
# selecting some countries
df_pivot=df_pivot[['India','China','Brazil']]
```

```python
df_pivot
```

 # 2 Data Visualization with Pandas

## 1 LinePlot

```python
df_pivot.plot(kind='line', xlabel='YEAR',ylabel='POPULATION',title='Population(1995-2020)',figsize=(8,4))
```

## 2 Bar plot

#### 2.1 Single Barplot

```python
# selecting only one year(2020)
df_pivot_2020=df_pivot[df_pivot.index.isin([2020])]
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020=df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar',color='orange',xlabel='YEAR',ylabel='POPULATION',title='Population 2020')
```

### 2.2 Barplot grouped by "n" variables

```python
# selecting a few years
df_pivot_sample=df_pivot[df_pivot.index.isin([1980,1990,2000,2010,2020])]
```

```python
# making grouped barplot
df_pivot_sample.plot(kind='bar',xlabel='YEAR',ylabel='POPULATION',title='Population (1980-2020)')
```

## 3 Piechart

```python
# changing column name
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
# making the plot
df_pivot_2020.plot(kind='pie',y='2020',title='Population in 2020')
```

## 4 Boxplot

### 4.1 single boxplot

```python
df_pivot['India'].plot(kind='box',color='green',ylabel='Population')
```

### 5.2 Multiple Boxplot

```python
df_pivot.plot(kind='box',color='green',ylabel='Population')
```

## 5 Histogram

### 5.1 For single country

```python
df_pivot['India'].plot(kind='hist')
```

### 5.2 For multiple country

```python
df_pivot[['India','China']].plot(kind='hist')
```
