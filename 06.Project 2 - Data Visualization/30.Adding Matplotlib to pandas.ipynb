```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('population_total.csv')
df
```

# 1 Making a Pivot table

.pivot() : returns reshaped Dataframe organized by given index/column values("pivot without aggregation")

```python
# dropping null values
df=df.dropna()
df
```

```python
# making a pivot table
df_pivot=df.pivot(index="year", columns="country",values="population")
```

```python
# selecting some countries
df_pivot=df_pivot[['India','China','Brazil']]
```

```python
df_pivot
```

# 1 LinePlot


```python
df_pivot.plot(kind='line', 
              xlabel='YEAR',
              ylabel='POPULATION',
              title='Population(1995-2020)',
              figsize=(8,4))

# save plot
plt.savefig('my_test.png')
# show plot
plt.show()
```
