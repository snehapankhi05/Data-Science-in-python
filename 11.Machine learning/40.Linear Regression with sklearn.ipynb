```python
import pandas as pd
```

```python
df=pd.read_csv('Boston House Prices.csv')
df
```

# 1 Linear Regression with scikit-learn

```python
from sklearn import linear_model
```

## 1.1 Define the dependent and independent variable

```python
# dependent variable
y=df['Value']

# independent variable
X=df[['Rooms','Distance']]
```

## 1.2 Fitting the model

```python
lm=linear_model.LinearRegression()
lm.fit(X,y)
```

## 1.3 Predicting values

```python
lm.predict(X)
```

## 1.4 regression Table

```python
# r2 score
lm.score(X,y)
```

```python
# coefficient
lm.coef_
```

```python
# intercept
lm.intercept_
```
