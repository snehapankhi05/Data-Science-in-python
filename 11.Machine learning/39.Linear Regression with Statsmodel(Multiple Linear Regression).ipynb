```python
df=pd.read_csv('Boston House Prices.csv')
df
```

# 1 Multiple Linear Regression

```python
import statsmodels.api as sm
```

## 1.1 Define the dependent and independent variable

```python
# dependent variable
y=df['Value']

# independent variable
X=df[['Rooms','Distance']]
```

## 1.2 Making the regression : Adding a constant and fitting the model

```python
# adding a constant
X=sm.add_constant(X)

# fitting the model
lm=sm.OLS(y,X).fit()
```

```python
lm.summary()
```
