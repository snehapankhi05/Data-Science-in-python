# 1 The re module

```python
import re
```

## 1.1 re.search()

```python
text = """Hello World.
I love python!!!!!.
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 123-456-123-123-456
"""
```

```python
# return the first match
re.search(r'\d',text)
```

```python
# flags
re.search(r'^hello',text,flags=re.I)
```

## 1.2 re.findall()

```python
# return all the matches
re.findall(r'\d',text)
```

# Excercise

```python
# match only the punctuation
text = """Hello World.
I love python!!!!!.
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 123-456-123-123-456
"""
re.findall(r'[^\w\s]',text)
```

```python
# match the right date format (MM-DD-YYYY)
date='''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}',date)
```

```python
# match the right username format (4-14 characters and only letters or digits)
user='''
username10
abc
10
'''
re.findall(r'[a-zA-Z0-9]{4,14}',user)
```
