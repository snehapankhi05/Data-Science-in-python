# 1 merge()

```python
import pandas as pd
df1=pd.DataFrame({'id':['A','B','C','D']
                 ,'age':[30,23,25,22]})
df2=pd.DataFrame({'id':['C','D','E','F']
                 ,'age':['Doctor','Accountant','Statistican','Developer']})
```

```python
df1
```

```python
df2
```

## 1.1 inner join

```python
df_inner=df1.merge(df2,on='id',how='inner')
```

```python
df_inner
```

### 1.1.1 Excercise

```python
# merge df_movies and df_ratings (inner join)
df_movies=pd.read_csv('IMDb movies.csv',low_memory=False)
df_ratings=pd.read_csv("IMDb ratings.csv",low_memory=False)

df_movies
```

```python
df_movies.merge(df_ratings,on='imdb_title_id',how='inner')
pd.merge(df_movies,df_ratings,on='imdb_title_id')
```

## 1.2 Outer join(full join)


```python
df1.merge(df2,on='id',how='outer')
```

### 1.2.1 Excercise

```python
# merge df_movies and merge df_rating 
df_movies.merge(df_ratings,on='imdb_title_id',how='outer')
```

## 1.3 (Exclusive) full join

```python
# indicator
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
# query()
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='left_only' or _merge=='right_only'")
```

### 1.3.1 Excercise

```python
df_movies.merge(df_ratings,on='imdb_title_id',how='outer',indicator=True)
```

```python
df_movies.merge(df_ratings,on='imdb_title_id',how='outer',indicator=True).query("_merge=='left_only' or _merge=='right_only' or _merge=='both'")
```

## 1.4 Left join

```python
df1.merge(df2,on='id',how='left')
```

### 1.4.1 Excercise

```python
# extract a 50% sample of the df_movies dataframe
df_movies_sample= df_movies.sample (frac=0.5)
df_movies_sample
```

```python
# merge df_movies_sample and df_ratings(left join)
df_movies_sample.merge(df_ratings,on='imdb_title_id',how='left')
```

## 1.5 Exclusive left join

```python
# indicator
df1.merge(df2,on='id',how='outer',indicator=True)
```

```python
# query()
df1.merge(df2,on='id',how='outer',indicator=True).query("_merge=='left_only' ")
```

### 1.5.1 Excercise

```python
# make copy of the df_movies dataframe
df_movies2=df_movies.copy()
df_movies2.loc[0,'imdb_title_id']
```

```python
# set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movies2.index:
    if index <1000:
        df_movies2.loc[index,'imdb_title_id']='tt1234567890'
df_movies2
```

```python
# merge df_movies2 and df_ratings (exclusive left join)
df_movies2.merge(df_ratings,on='imdb_title_id',how='outer',indicator=True).query("_merge=='left_only'")
```

## 1.6 Right join


```python
df1.merge(df2,on='id',how='right')
```

```python
# extract a 50% sample of the df_ratings dataframe
df_ratings_sample= df_ratings.sample (frac=0.3)
df_ratings_sample
```

```python
# merge df_movies and df_ratings sample(right join)
df_movies.merge(df_ratings_sample,on='imdb_title_id',how='right')
```
